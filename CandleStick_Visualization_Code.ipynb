```python
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio
```

```python
btcdf = pd.read_csv('BITSTAMP_BTCUSD_1D.csv')
ethdf = pd.read_csv('COINBASE_ETHUSD_1D.csv')
coindf = pd.read_csv('COINBASE_LINKUSD_1D.csv')
```

```python
import plotly.graph_objs as go
import plotly.io as pio
import numpy as np
import pandas as pd

def plot_candlestick_charts(dataframes, dataframe_names, output_file=None):
    """ Plot candlestick charts for multiple DataFrames. Args: - dataframes (list): List of DataFrames containing OHLC data. - dataframe_names (list): List of names for each DataFrame to be used as labels. - output_file (str, optional): Output file path to save the plot as an HTML file. Default is None. """
    # Calculate maximum values for each DataFrame
    max_values = [df[['open', 'high', 'low', 'close']].max().max() for df in dataframes]
    # Generate colors for each DataFrame
    colors = [{'increasing': f'rgb({np.random.randint(0, 256)}, {np.random.randint(0, 256)}, {np.random.randint(0, 256)})',
               'decreasing': f'rgb({np.random.randint(0, 256)}, {np.random.randint(0, 256)}, {np.random.randint(0, 256)})'}
              for _ in range(len(dataframes))]
    # Create traces for each DataFrame
    traces = []
    for i, (df, name) in enumerate(zip(dataframes, dataframe_names)):
        # Convert Unix timestamps to datetime
        df.index = pd.to_datetime(df['time'], unit='s')
        trace = go.Candlestick(x=df.index,
                                open=df['open'] / max_values[i],
                                high=df['high'] / max_values[i],
                                low=df['low'] / max_values[i],
                                close=df['close'] / max_values[i],
                                name=name,
                                increasing_line_color=colors[i]['increasing'],
                                decreasing_line_color=colors[i]['decreasing'])
        traces.append(trace)
    # Create a figure with the candlestick traces
    fig = go.Figure(data=traces)
    # Update layout
    fig.update_layout(title='Normalized Candlestick Chart',
                      xaxis_title='Time',
                      yaxis_title='Price (Normalized)')
    # Show the plot
    if output_file:
        pio.write_html(fig, output_file)
    else:
        pio.show(fig)
```

```python
# Example usage
dataframes = [btcdf, ethdf, coindf]
dataframe_names = ['BTC', 'ETH', 'Coin']
plot_candlestick_charts(dataframes, dataframe_names)

#This is if you want a full interactable page of the graph
#plot_candlestick_charts(dataframes, dataframe_names, output_file='candlestick_charts.html')
```
